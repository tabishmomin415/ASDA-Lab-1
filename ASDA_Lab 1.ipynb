```python
import string
from collections import defaultdict
import pandas as pd
```

```python
common_nouns = ["idea", "reaction", "term", "France", "Amar", "mid", "Limit", "degrowth", "capitalism", 
                "agenda", "model", "process", "methodology", "ecroissance", "GDP", "economics", "Georgesscu-Roegen"]

def is_noun(word):
    if word in common_nouns:
        return True
    if word[0].isupper():
        return True
    if word.endswith(("tion", "ment", "ism", "ity")):
        return True
    return False
```

```python
def process_string(text):
    word_info = defaultdict(lambda: {
        "frequency": 0, 
        "type": "", 
        "length": 0, 
        "lowercase": "", 
        "only_a_z": False,
        "is_noun": False
    })
    
    allowed_chars = string.ascii_letters + "'"
    cleaned_text = ''.join([char if char in allowed_chars or char.isspace() else ' ' for char in text])
    
    words = cleaned_text.split()
    
    for word in words:

        lowercase_word = word.lower()

        only_a_z = all(char.isalpha() or char == "'" for char in lowercase_word)

        is_noun_flag = is_noun(word)

        word_info[word]['frequency'] += 1
        word_info[word]['type'] = "alpha" if only_a_z else "non-alpha"
        word_info[word]['length'] = len(word)
        word_info[word]['lowercase'] = lowercase_word
        word_info[word]['only_a_z'] = only_a_z
        word_info[word]['is_noun'] = is_noun_flag
    
    return word_info
```

```python
word_data = process_string(input_string)
```

```python
table_data = {
    "Word": [],
    "Frequency": [],
    "Type": [],
    "Length": [],
    "Lowercase": [],
    "Only a-z": [],
    "Is Noun": []
}

for word, info in word_data.items():
    table_data["Word"].append(word)
    table_data["Frequency"].append(info["frequency"])
    table_data["Type"].append(info["type"])
    table_data["Length"].append(info["length"])
    table_data["Lowercase"].append(info["lowercase"])
    table_data["Only a-z"].append(info["only_a_z"])
    table_data["Is Noun"].append(info["is_noun"])
```
